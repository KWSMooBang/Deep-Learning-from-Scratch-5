### Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

x_means = []
N = 1 

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)

# 그래프 그리기
plt.hist(x_means, bins='auto', density=True)
plt.title(f" N = {N}")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

```python
def sample(N=1):
    x_means = []
    for _ in range(10000):
        xs = []
        for n in range(N):
            x = np.random.rand()
            xs.append(x)
        mean = np.mean(xs)
        x_means.append(mean)
    return x_means

x_means_1 = sample(1)
x_means_2 = sample(2)
x_means_4 = sample(4)
x_means_10 = sample(10)


fig, axes = plt.subplots(2, 2)

axes[0, 0].hist(x_means_1, bins='auto', density=True)
axes[0, 0].set_title(f" N = 1 ")
axes[0, 0].set_ylabel("Probability Density")
axes[0, 0].set_xlim(-0.05, 1.05)
axes[0, 0].set_ylim(0, 5)

axes[0, 1].hist(x_means_2, bins='auto', density=True)
axes[0, 1].set_title(f" N = 2 ")
axes[0, 1].set_ylim(0, 5)
axes[0, 1].set_xlim(-0.05, 1.05)

fig.subplots_adjust(hspace=0.5)

axes[1, 0].hist(x_means_4, bins='auto', density=True)
axes[1, 0].set_title(f" N = 4 ")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("Probability Density")
axes[1, 0].set_xlim(-0.05, 1.05)
axes[1, 0].set_ylim(0, 5)

axes[1, 1].hist(x_means_10, bins='auto', density=True)
axes[1, 1].set_title(f" N = 10 ")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_xlim(-0.05, 1.05)
axes[1, 1].set_ylim(0, 5)
```
